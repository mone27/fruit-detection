```python
import os
from pathlib import Path
from random import randint
import cv2
import matplotlib.pyplot as plt
from classic_segmentation import ClassicSegmentation
from PIL import Image
```

```python

all_folders = ['Cetriolo', 'Peperoncino', 'Patata', 'pomodoro_ciliegino', 'Banana', 'Carota', 'Albicocca',
              'Kiwi', 'Zucchino', 'Pera', 'Mela', 'Lime', 'Champignon',
              'Prugna rossa', 'Pesca', 'Cipolla', 'Pomodoro_sanmarzano']
root_folder = Path("./Dataset")
ext = '.tif'  # source file extention
ext_save = '.png'  # save file extention
root_folder_save = Path('dataset_class')
```

```python
def save_image(filename, image):
    
    cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    print(f"filename: '{filename}'")
    img = Image.fromarray(image)
    img.save(filename, "PNG")
    #success = cv2.imwrite(str(filename), image)
    
    #if success: print(f"saved: {filename}")
    #else:
        #raise Exception(f"cannot save {filename}")
    
    
def save_folder(folder): 
    # create folder
    

    for img_path in (root_folder / folder).glob(f"*{ext}"):
        print(img_path)
        for image_counter, image in enumerate(ClassicSegmentation(img_path).small_images):
            # generate a random number to make ~ 20% of images in test file
            # define filename for test or for train
            if randint(0,9) < 8:
                folder_train = "train"
            else:
                folder_train = "test"
            print(folder_train)
            filename = root_folder_save / str(folder_train) / str(folder) / f"{str(img_path.name)[:-len(ext)]}_{image_counter}{ext_save}"
            filename.parent.mkdir(parents=True, exist_ok=True)       
            save_image(filename, image)
                

def save_all():
    for folder in all_folders:
        save_folder(folder)
```

```python
save_all()
```
